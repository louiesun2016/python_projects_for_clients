```python
#latitude 20.32 25.52
#longitude 109.75  117.33
min_longitude = 113.06
max_longitude = 113.7
min_latitude = 23.1
max_latitude = 23.6

sites['longitude'] = pd.to_numeric(sites['longitude'], errors='coerce')
sites['latitude'] = pd.to_numeric(sites['latitude'], errors='coerce')

filtered_sites = sites[(sites['longitude'] >= min_longitude) & (sites['longitude'] <= max_longitude) & 
                 (sites['latitude'] >= min_latitude) & (sites['latitude'] <= max_latitude)]
```

```python
SO2.set_index('date', inplace=True)
NO2.set_index('date', inplace=True)
CO.set_index('date', inplace=True)
PM25.set_index('date', inplace=True)
PM10.set_index('date', inplace=True)
O3.set_index('date', inplace=True)
#m = SO2.groupby(pd.Grouper(freq = 'Y')).mean(skipna=True)
```

```python
SO2.drop('type', axis=1, inplace=True)
NO2.drop('type', axis=1, inplace=True)
CO.drop('type', axis=1, inplace=True)
PM25.drop('type', axis=1, inplace=True)
PM10.drop('type', axis=1, inplace=True)
O3.drop('type', axis=1, inplace=True)
```

```python
O3
```

```python
#SO2.resample('Y').mean(skipna=True)
annual_mean_SO2 = SO2.resample('Y').agg(pd.Series.mean, skipna=True)
annual_mean_NO2 = NO2.resample('Y').agg(pd.Series.mean, skipna=True)
annual_mean_CO = CO.resample('Y').agg(pd.Series.mean, skipna=True)
annual_mean_PM25 = PM25.resample('Y').agg(pd.Series.mean, skipna=True)
annual_mean_PM10 = PM10.resample('Y').agg(pd.Series.mean, skipna=True)
annual_mean_O3 = O3.resample('Y').agg(pd.Series.mean, skipna=True)
# SO2['date']
```

```python
annual_mean_SO2
annual_mean_NO2
```

```python
annual_mean_SO2_all = annual_mean_SO2.mean(axis=1)
annual_mean_CO_all = annual_mean_CO.mean(axis=1)
annual_mean_NO2_all = annual_mean_NO2.mean(axis=1)
annual_mean_PM25_all = annual_mean_PM25.mean(axis=1)
annual_mean_PM10_all = annual_mean_PM10.mean(axis=1)
annual_mean_O3_all = annual_mean_O3.mean(axis=1)
```

```python
annual_mean_CO_all.values
```

```python
import matplotlib.pyplot as plt 
import numpy as np
 
# set width of bar 
barWidth = 0.1
fig = plt.subplots(figsize =(16, 8)) 
 
 
# Set position of bar on X axis 
br1 = np.arange(len(annual_mean_SO2_all)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 
br4 = [x + barWidth for x in br3] 
br5 = [x + barWidth for x in br4] 
br6 = [x + barWidth for x in br5] 
 
# Make the plot
plt.bar(br1, annual_mean_SO2_all.values, color ='w', width = barWidth, 
        edgecolor ='k', hatch = '....',label ='SO₂') 
plt.bar(br2, annual_mean_CO_all.values*10.0, color ='w', width = barWidth, 
        edgecolor ='k', hatch = '////',label ='CO x 10$^{-2}$') 
plt.bar(br3, annual_mean_NO2_all.values, color ='w', width = barWidth, 
        edgecolor ='k', hatch = '+++',label ='NO₂') 
plt.bar(br4, annual_mean_O3_all.values, color ='w', width = barWidth, 
        edgecolor ='k', hatch = '\\\\',label ='O₃') 
plt.bar(br5, annual_mean_PM25_all.values, color ='w', width = barWidth, 
        edgecolor ='k', hatch = 'ooo',label ='PM2.5') 
plt.bar(br6, annual_mean_PM10_all.values, color ='w', width = barWidth, 
        edgecolor ='k', hatch = '***',label ='PM10') 
 
# Adding Xticks 
plt.xlabel('年份', fontweight ='bold', fontsize = 15, font='SimSun') 
plt.ylabel('污染物浓度/($\mu$g/m$^3$)', fontweight='bold', fontsize=15, font='SimSun')
plt.xticks([r + barWidth for r in range(len(annual_mean_SO2_all))], 
        ['2018', '2019', '2020', '2021', '2022'])
plt.ylim(0,100)
 
plt.legend()
plt.savefig('air_quality_plot.png')
plt.show() 
```
